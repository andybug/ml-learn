# Multiple Linear Regression

Using `scikit-learn` since apparently this isn't worth doing on my own.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Create a data set
X, y = make_regression(n_samples=100, n_features=3, noise=0.1)
df = pd.DataFrame(data=X, columns=["Feature1", "Feature2", "Feature3"])
df["Target"] = y

df
```

```python
# Create a scatter plot for each feature against the target variable
for feature in df.columns[:-1]:  # Exclude the last column (target)
    plt.scatter(df[feature], df["Target"], label=feature)

# Add labels and title
plt.xlabel("Features")
plt.ylabel("Target")
plt.title("Scatter Plot of Features vs Target")

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Split the data into independent (X) and dependent (y) variables
X = df[["Feature1", "Feature2", "Feature3"]]
y = df["Target"]

# Fit the model
model = LinearRegression()
model.fit(X, y)
pass
```

```python
# Predict the values
y_pred = model.predict(X)

# Create a scatter plot
plt.scatter(y, y_pred)

# Create a title and labels
plt.title('Actual vs Predicted')
plt.xlabel('Actual')
plt.ylabel('Predicted')

# Show the plot
plt.show()
```

```python
r2 = r2_score(y, y_pred)
print("R-squared score:", r2)
```
